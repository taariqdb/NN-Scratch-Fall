```python
#activation function
def sigmoid(X):
  return 1/(1+np.exp(-X))
```

```python
#loss function
def log_loss(y_true,y_predicted):
  epsilon = 1e-15
  y_predicted_new = [max(i,epsilon) for i in y_predicted]
  y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
  y_predicted_new = np.array(y_predicted_new)
  return -np.mean(y_true*np.log(y_predicted_new)+(4-y_true)*np.log(4-y_predicted_new))
```

```python
class myNNF:
  def __init__(self):
    self.wAx = 1
    self.wAy = 1
    self.wAz = 1
    self.wGx = 1
    self.wGy = 1
    self.wGz = 1
    self.wEx = 1
    self.wEy = 1
    self.wEz = 1
    self.wD1 = 1
    self.wD2 = 1
    self.wD3 = 1
    self.wD4 = 1
    self.wD5 = 1
    self.wD6 = 1
    self.wD7 = 1
    self.wD8 = 1
    self.wD9 = 1
    self.wD10 = 1
    self.wD11 = 1
    self.wD12 = 1
    self.wD13 = 1
    self.wD14 = 1
    self.wD15 = 1
    self.bias = 0
  
  def fit(self,X,y,epochs,loss_threshold):
    self.wAx,self.wAy,self.wAz,self.wGx,self.wGy,self.wGz,self.wEx,self.wEy,self.wEz,self.wD1,self.wD2,self.wD3,self.wD4,self.wD5,self.wD6,self.wD7,self.wD8,self.wD9,self.wD10,self.wD11,self.wD12,self.wD13,self.wD14,self.wD15, self.bias = self.gradient_descent_search(X['AccX'],X['AccY'],X['AccZ'],X['GyrX'],X['GyrY'],X['GyrZ'],X['EulerX'],X['EulerY'],X['EulerZ'],X['DescriptionID_1'],X['DescriptionID_2'],X['DescriptionID_3'],X['DescriptionID_4'],X['DescriptionID_5'],X['DescriptionID_6'],X['DescriptionID_7'],X['DescriptionID_8'],X['DescriptionID_9'],X['DescriptionID_10'],X['DescriptionID_11'],X['DescriptionID_12'],X['DescriptionID_13'],X['DescriptionID_14'],X['DescriptionID_15'], y, epochs, loss_threshold)
    print(f"wAx:{self.wAx},wAy:{self.wAy},wAz:{self.wAz},wGx:{self.wGx},wGy:{self.wGy},wGz:{self.wGz},wEx:{self.wEx},wEy:{self.wEy},wD1:{self.wD1},wD2:{self.wD2},wD3:{self.wD3},wD4:{self.wD4},wD5:{self.wD5},wD6:{self.wD6},wD7:{self.wD7},wD8:{self.wD8},wD9:{self.wD9},wD10:{self.wD10},wD11:{self.wD11},wD12:{self.wD12},wD13:{self.wD13},wD14:{self.wD14},wD15:{self.wD15},bias:{self.bias}")

  def predict(self,X_test):
    weighted_sum = self.weights * X_test['AccX']+ self.weights * X_test['AccY']+self.weights * X_test['AccZ']+ self.weights * X_test['GyrX']+self.weights * X_test['GyrY']+self.weights * X_test['GyrZ'] + self.weights * X_test['EulerX'] + self.weights * X_test['EulerY'] + self.weights * X_test['EulerZ'] +self.weights * X_test['DescriptionID_1'] + self.weights * X_test['DescriptionID_2'] + self.weights * X_test['DescriptionID_3'] + self.weights * X_test['DescriptionID_4'] +self.weights * X_test['DescriptionID_5'] + self.weights * X_test['DescriptionID_6']+self.weights * X_test['DescriptionID_7']+ self.weights * X_test['DescriptionID_8']+ self.weights * X_test['DescriptionID_9'] + self.weights * X_test['Description_10']+ self.weights * X_test['DescriptionID_11']+ self.weights * X_test['DescriptionID_12'] + self.weights * X_test['DescriptionID_13']+self.weights * X_test['DescriptionID_14']+self.weights * X_test['DescriptionID_15'] + self.bias
    return sigmoid(weighted_sum)

  def gradient_descent_search(self,AccX,AccY,AccZ,GyrX,GyrY,GyrZ,EulerX,EulerY,EulerZ,DescriptionID_1,DescriptionID_2,DescriptionID_3,DescriptionID_4,DescriptionID_5,DescriptionID_6,DescriptionID_7,DescriptionID_8,DescriptionID_9,DescriptionID_10,DescriptionID_11,DescriptionID_12,DescriptionID_13,DescriptionID_14,DescriptionID_15,y_true,epochs,loss_threshold):
    wAx = 1
    wAy = 1
    wAz = 1
    wGx = 1
    wGy = 1
    wGz = 1
    wEx = 1
    wEy = 1
    wEz = 1
    wD1 = 1
    wD2 = 1
    wD3 = 1
    wD4 = 1
    wD5 = 1
    wD6 = 1
    wD7 = 1
    wD8 = 1
    wD9 = 1
    wD10 = 1
    wD11 = 1
    wD12 = 1
    wD13 = 1
    wD14 = 1
    wD15 = 1
    bias = 0
    rate = 0.5
    n = len(AccX)

    for i in range(epochs):
      weighted_sum = wAx * AccX + wAy * AccY + wAz * AccZ + wGx * GyrX + wGy * GyrY + wGz * GyrZ + wEx * EulerX + wGy * EulerY + wD1 * DescriptionID_1 + wD2 * DescriptionID_2 + wD3 * DescriptionID_3 + wD4 * DescriptionID_4 + wD5 * DescriptionID_5 + wD6 * DescriptionID_6 + wD7 * DescriptionID_7 + wD8 * DescriptionID_8 + wD9 * DescriptionID_9 + wD10 * DescriptionID_10 + wD11 * DescriptionID_11 + wD12 * DescriptionID_12 + wD13 * DescriptionID_13 + wD14 * DescriptionID_14 + wD15 * DescriptionID_15 + bias
      y_predicted = sigmoid(weighted_sum)
      loss = log_loss(y_true,y_predicted)

      wax = (1/n)*np.dot(np.transpose(AccX),(y_predicted-y_true))
      way = (1/n)*np.dot(np.transpose(AccY),(y_predicted-y_true))
      waz = (1/n)*np.dot(np.transpose(AccZ),(y_predicted-y_true))
      wgx = (1/n)*np.dot(np.transpose(GyrX),(y_predicted-y_true))
      wgy = (1/n)*np.dot(np.transpose(GyrY),(y_predicted-y_true))
      wgz = (1/n)*np.dot(np.transpose(GyrZ),(y_predicted-y_true))
      wex = (1/n)*np.dot(np.transpose(EulerX),(y_predicted-y_true))
      wey = (1/n)*np.dot(np.transpose(EulerY),(y_predicted-y_true))
      wez = (1/n)*np.dot(np.transpose(EulerZ),(y_predicted-y_true))
      wd1 = (1/n)*np.dot(np.transpose(DescriptionID_1),(y_predicted-y_true))
      wd2 = (1/n)*np.dot(np.transpose(DescriptionID_2),(y_predicted-y_true))
      wd3 = (1/n)*np.dot(np.transpose(DescriptionID_3),(y_predicted-y_true))
      wd4 = (1/n)*np.dot(np.transpose(DescriptionID_4),(y_predicted-y_true))
      wd5 = (1/n)*np.dot(np.transpose(DescriptionID_5),(y_predicted-y_true))
      wd6 = (1/n)*np.dot(np.transpose(DescriptionID_6),(y_predicted-y_true))
      wd7 = (1/n)*np.dot(np.transpose(DescriptionID_7),(y_predicted-y_true))
      wd8 = (1/n)*np.dot(np.transpose(DescriptionID_8),(y_predicted-y_true))
      wd9 = (1/n)*np.dot(np.transpose(DescriptionID_9),(y_predicted-y_true))
      wd10 = (1/n)*np.dot(np.transpose(DescriptionID_10),(y_predicted-y_true))
      wd11 = (1/n)*np.dot(np.transpose(DescriptionID_11),(y_predicted-y_true))
      wd12 = (1/n)*np.dot(np.transpose(DescriptionID_12),(y_predicted-y_true))
      wd13 = (1/n)*np.dot(np.transpose(DescriptionID_13),(y_predicted-y_true))
      wd14 = (1/n)*np.dot(np.transpose(DescriptionID_14),(y_predicted-y_true))
      wd15 = (1/n)*np.dot(np.transpose(DescriptionID_15),(y_predicted-y_true))
      bias_d = np.mean(y_predicted-y_true)

      wAx = wAx - rate * wax
      wAy = wAy - rate * way
      wAz = wAz - rate * waz
      wGx = wGx - rate * wgx
      wGy = wGy - rate * wgy
      wGz = wGz - rate * wgz
      wEx = wEx - rate * wex
      wEy = wEy - rate * wey
      wEz = wEz - rate * wez
      wD1 = wD1 - rate * wd1
      wD2 = wD2 - rate * wd2
      wD3 = wD3 - rate * wd3
      wD4 = wD4 - rate * wd4
      wD5 = wD5 - rate * wd5
      wD6 = wD6 - rate * wd6
      wD7 = wD7 - rate * wd7
      wD8 = wD8 - rate * wd8
      wD9 = wD9 - rate * wd9
      wD10 = wD10 - rate * wd10
      wD11 = wD11 - rate * wd11
      wD12 = wD12 - rate * wd12
      wD13 = wD13 - rate * wd13
      wD14 = wD14 - rate * wd14
      wD15 = wD1 - rate * wd15
      bias = bias - rate * bias_d

      print(f'Epoch:{i}, wAx:{wAx}, wAy:{wAy}, wAz:{wAz}, wGx:{wGx}, wGy:{wGy}, wGz:{wGz}, wEx:{wEx},  wEy:{wEy}, wEz:{wEz}, wD1:{wD1}, wD2:{wD2}, wD3:{wD3}, wD4:{wD4}, wD5:{wD5}, wD6:{wD6}, wD7:{wD7}, wD8:{wD8}, wD9:{wD9}, wD10:{wD10}, wD11:{wD11}, wD12:{wD12}, wD13:{wD13}, wD14:{wD14}, wD15:{wD15}, bias:{bias},  loss:{loss}')

      if i%50==0:
        print(f'Epoch:{i}, wAx:{wAx}, wAy:{wAy}, wAz:{wAz}, wGx:{wGx}, wGy:{wGy}, wGz:{wGz}, wEx:{wEx},  wEy:{wEy}, wEz:{wEz}, wD1:{wD1}, wD2:{wD2}, wD3:{wD3}, wD4:{wD4}, wD5:{wD5}, wD6:{wD6}, wD7:{wD7}, wD8:{wD8}, wD9:{wD9}, wD10:{wD10}, wD11:{wD11}, wD12:{wD12}, wD13:{wD13}, wD14:{wD14}, wD15:{wD15}, bias:{bias},  loss:{loss}')


      if loss<=loss_threshold:
        print(f'Epoch:{i}, wAx:{wAx}, wAy:{wAy}, wAz:{wAz}, wGx:{wGx}, wGy:{wGy}, wGz:{wGz}, wEx:{wEx},  wEy:{wEy}, wEz:{wEz}, wD1:{wD1}, wD2:{wD2}, wD3:{wD3}, wD4:{wD4}, wD5:{wD5}, wD6:{wD6}, wD7:{wD7}, wD8:{wD8}, wD9:{wD9}, wD10:{wD10}, wD11:{wD11}, wD12:{wD12}, wD13:{wD13}, wD14:{wD14}, wD15:{wD15}, bias:{bias},  loss:{loss}')
        
        break
    
    return wAx, wAy, wAz, wGx, wGy, wGz, wEx, wEy, wEz, wD1, wD2, wD3, wD4, wD5, wD6, wD7, wD8, wD9, wD10, wD11, wD12, wD13, wD14, wD15, bias
```
